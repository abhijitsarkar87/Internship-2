## Task 2: View the distributions of the various features in the data set and calculate their central tendencies.

- We will now look at the distributions of the various features in the data set
- We will also calculate appropriate measures of central tendency for these features

```python
# create histogram of the "Fixed acidity" features

plt.figure(figsize=(6,4))

sns.histplot(data=data,x='fixed acidity',color='orange',edgecolor='linen',alpha=0.9,bins=5)# alpha will give itensity of bar color

plt.title("HISTOGRAM OF FIXED ACIDITY")
plt.xlabel('Fixed Acidity')
plt.ylabel("Count")
plt.show()
```

**Observation**

- We observe that the histogram is normally distributed.
- The maximum count of values for fixed acidity lies between 6 to 8

Lets see the measure of central tendency in working!

    1. Mean
    2. Median
    3. Mode

### Calculate Mean

```python
data['fixed acidity'].mean()
```

### Calculate Median

```python
data['fixed acidity'].median()
```

#### Now we will plot the Mean and Median on the graph 

```python
plt.figure(figsize=(11,6)) # expanded the graph to see the clear mean and median in the graph
sns.histplot(data=data,x='fixed acidity',color='orange',edgecolor='black',alpha=0.9,bins=5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel("Count")
plt.vlines(data['fixed acidity'].mean(),colors='purple',ymin=0,ymax=4000,label='Mean')
plt.vlines(data['fixed acidity'].median(),colors='brown',ymin=0,ymax=4000,label='Median')
plt.legend(loc=5)
plt.show()
```

### Volatile Acidity

```python
plt.figure(figsize=(11,6))

sns.histplot(data=data,x='volatile acidity',color='green',bins=5,edgecolor='black',alpha=0.9)

plt.title("Histogram of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Counts")
plt.vlines(data['volatile acidity'].mean(),ymin=0,ymax=3500,color='red',label="MEAN")
plt.vlines(data['volatile acidity'].median(),ymin=0,ymax=3500,color='brown',label="MEDIAN")
plt.legend(loc=4)
plt.show()
```

### OBSERVATION

- We observe that this histogram is not well distributed, It is skewed a little towards the right (Mean is greater than Medin). 
- We have seen skewness, therefore we can check the distribution using distplot function.

```python
#Mean
data['volatile acidity'].mean()
```

```python
# Median
data['volatile acidity'].median()
```

```python
plt.figure(figsize=(11,6))

sns.distplot(data['volatile acidity'],color='blue')

plt.title("Histogram of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Counts")
plt.show()
```

```python
data['volatile acidity'].skew() # positive skewed
```

- We can clearly see that the skewness value is greater than 1, hence it is positively skewed.

**Left OR RIGHT SKEWED** 

A distribution is left skewed if it has a 'tail' on the left side of the distribution.Left Skewed Distribution: Mean<Median < Mode


**Right OR NEGATIVE SKEWED**

A distribution is right skewed if it has a tail on the right side of the distribution
           *Right Skewed Distribution: Mode < Median < Mean* 

# Skewness

Some distributions of data, such as the bell curve or normal distribution, are symmetric. This means that the left of the distribution are perfect mirror image of one another. Not every distribution of data is symmetric. Sets of data are not symmetric are said to be **asymmetric** The measure of how asymmetric a distribution can be is called as skewness.


The mean, median and mode are all measures of the center of a set of data. The skewness of the data can be determined by how these quantities are related to one another.


   1. If Skewness(S)=0 then the distribution is normal distribution. ie. No Skew: Mean = Median = Mode
   2. If Skewness(S)>0 then the distribution is positive skewed.
   3. If Skewness(S)<0 then the distribution is negative skewed.
